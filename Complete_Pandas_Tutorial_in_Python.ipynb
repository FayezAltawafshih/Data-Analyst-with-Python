Pandas DataFrame:

Creating a Pandas Dataframe

```python
# imborting the boston house price data
from sklearn.datasets import load_iris
```

```python
iris_dataset = load_iris()
```

```python
type(iris_dataset)
```

```python
diabetes_df.head()
```

```python
# Locating a particular column
print(iris_df.iloc[:,0]) # first columns
print(iris_df.iloc[:,1]) # Second columns
print(iris_df.iloc[:,2]) # Third Columns
print(iris_df.iloc[:,-1]) # last Columns
```
